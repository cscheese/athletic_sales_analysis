```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# define the file paths
sales_2020_path = Path('Resources/athletic_sales_2020.csv')
sales_2021_path = Path('Resources/athletic_sales_2021.csv')

# Read the CSV files into DataFrames.
sales_2020 = pd.read_csv(sales_2020_path)
sales_2021 = pd.read_csv(sales_2021_path)
```

```python
# Display the 2020 sales DataFrame
display(sales_2020.head(5))
display(sales_2020.tail(5))
```

```python
# Display the 2021 sales DataFrame
#from IPython.display: sales_2021
display(sales_2021.head(5))
display(sales_2021.tail(5))
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames using an inner join and reset the index.
combined_sales = pd.concat([sales_2020, sales_2021], axis=0, join='inner', ignore_index=True)
combined_sales.reset_index(drop=True, inplace=True)

# Display the first 5 rows and last 5 rows of the combined DataFrame.
display(combined_sales.head(5))
display(combined_sales.tail(5))
```

```python
# Check if any values are null.
null_values = combined_sales.isnull().sum()
print(null_values)
```

```python
# Check the data type of each column
combined_sales.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
df = combined_sales
df['invoice_date'] = pd.to_datetime(df['invoice_date'], format='%m/%d/%y')
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
combined_sales.columns
```

```python
# Show the number products sold for region, state, and city.
sales_region_state_city = combined_sales.groupby(['region', 'state','city'])['units_sold'].sum().reset_index()

# Rename the sum to "Total_Products_Sold".
sales_region_state_city.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

top_5_results = sales_region_state_city.sort_values(by='Total_Products_Sold', ascending=False)

# Show the top 5 results.
top_5_results.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
#combined_sales = pd.DataFrame

# Create a pivot table to show the number of products sold for region, state, and city.
sales_pivot = combined_sales.pivot_table(index=['region', 'state', 'city'], values='units_sold', aggfunc='sum').reset_index()

# Rename the sum to "Total_Products_Sold".
sales_pivot.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Sort the results to show the top 5 regions, states, and cities by total products sold.
top_5_results = sales_pivot.sort_values(by='Total_Products_Sold', ascending=False).head()

# Show the top 5 results.
top_5_results
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
df = womens_footwear
pivot_table = df.pivot_table(index=['invoice_date'], values='total_sales', aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the table.
pivot_table.head(10)

# I created this code by using Xpert, copying some code from above, and generating some of the code on my own. 
# Initially I used the wrong df - reverting back to the "combined_sales", but in comparing the output, I discovered my error.
# I removed the reset index that had been a part of the copied code from above and set the header to display 10 as in the output example. 
# I'm finally feeling like I am understanding how this all fits together. 
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
pivot_table.index = pd.to_datetime(pivot_table.index)
resampled_pivot_table = pivot_table.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales"
resampled_pivot_table.sort_values(by='Total Sales', ascending=False, inplace=True)

# Show the resampled pivot table
resampled_pivot_table.head(10)

# I used Xpert and prior code and critical thinking to generate this code and output. 
# note:the initial code recommended by Xpert had errors in it. 
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
pivot_table.index = pd.to_datetime(pivot_table.index)
resampled_pivot_table = pivot_table.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales"
resampled_pivot_table.sort_values(by='Total Sales', ascending=False, inplace=True)

# Show the resampled pivot table
resampled_pivot_table.head(10)

# I copied the above code and then revised it on my own - recalling from lecture the meaning of the D & W in datetime. Yea!!
```
